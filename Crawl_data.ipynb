```python
import requests
from bs4 import BeautifulSoup

# Đường dẫn URL của trang web cần lấy thông tin
website = 'https://www.chotot.com/mua-ban-laptop?page='
href_list =[]
partner_list =[]

for i in range(1, 100):
    url = website + str(i)
    print(url)
# Lấy nội dung của trang web
    response = requests.get(url)

# Kiểm tra xem request có thành công không
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        li_tags = soup.find_all('li', class_='AdItem_wrapperAdItem__S6qPH')

        # Lặp qua các phần tử con và lưu thông tin vào bảng
        for li_tag in li_tags:
            a_tag = li_tag.find('a', class_='AdItem_adItem__gDDQT')
            href= a_tag.get('href') if a_tag else None
            href_list.append(href)

            div_protected_label = li_tag.find('div', class_='AdBody_protectedLabelInTitle__HQgj6')

            partner = 'yes' if div_protected_label else "No"
            partner_list.append(partner)

    else: print(f'fail:{url}')
```

```python
import pandas as pd
df= pd.DataFrame({'href':href_list, 'Doi tac': partner_list})
```

```python
df.to_csv('Link.csv', index = False)
```

```python
import csv
import requests
from bs4 import BeautifulSoup

# Khởi tạo danh sách chứa các thuộc tính itemprop cần tìm
itemprops = ['pc_brand', 'pc_model', 'elt_condition', 'laptop_screen_size', 'elt_warranty', 'pc_cpu', 'pc_ram', 'pc_drive_capacity',
             'elt_origin', 'pc_vga', 'price']

# Khởi tạo danh sách chứa dữ liệu
data_list = []

# Mở tệp CSV và đọc các URL
with open('full-link.csv', newline='') as csvfile:  # Thay thế 'tệp.csv' bằng tên tệp CSV của bạn
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        url = row[0]  # Giả sử cột đầu tiên chứa các URL

        # Gửi yêu cầu GET đến trang web
        response = requests.get(url)

        # Kiểm tra xem request có thành công không
        if response.status_code == 200:
            # Phân tích cú pháp HTML bằng BeautifulSoup
            soup = BeautifulSoup(response.content, 'html.parser')

            # Khởi tạo dictionary để lưu trữ thông tin của mỗi URL
            data = {}

            # Tìm và lưu thông tin từ các thuộc tính itemprop cần tìm
            for itemprop in itemprops:
                item = soup.find(attrs={"itemprop": itemprop})
                if item:
                    data[itemprop] = item.get_text()
                else:
                    data[itemprop] = ''

            # Thêm dictionary vào danh sách
            data_list.append(data)

        else:
            print(f"Yêu cầu không thành công cho URL: {url}")

# Ghi thông tin vào tệp CSV
with open('output.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=itemprops)
    writer.writeheader()
    for data in data_list:
        writer.writerow(data)
```
